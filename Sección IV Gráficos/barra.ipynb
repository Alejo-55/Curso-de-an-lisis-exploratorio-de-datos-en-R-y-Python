## Gráfico de barras

A través de la librería MatplotLib se pueden crear múltiples tipos de gráficos uno de los usados es la gráfica de barras, las gráficas de barras suelen ser usadas para visualizar el valor de datos categóricos. Por un lado, la gráfica o gráfico de barras tiene barras rectangulares con longitudes proporcionales a los valores y ejes que representan. Las gráficas de barras se utilizan para comparar dos o más valores. Las barras pueden ser horizontales o verticales.

### Gráfico de barras básico
Claro que en Python existen funciones que nos eviten tener que escribir los nombres de las barras y sus valores en el código. Sin embargo, en este caso se está haciendo así para aprender a crear un gráfico desde ceros, es decir, si no tuviéramos ninguna base de datos abierta, cómo podríamos hacer un gráfico: esto es posible poniendo los nombres de las barras y sus respectivos valores, tal como se muestra acá:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Continentes = ['África', 'Asia', 'Europa', 'Américas', 'Oceanía']
Observaciones = [624,396,360,300,24]
ax.bar(Continentes,Observaciones)
plt.show()
```

### Cambiando los colores de las barras

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Continentes = ['África', 'Asia', 'Europa', 'Américas', 'Oceanía']
Observaciones = [624,396,360,300,24]
ax.bar(Continentes,Observaciones, color=['orange', 'red', 'green', 'blue', 'cyan'])
plt.show()
```

También es posible dejar las barras con un color neutro o con un color único a su alrededor, además de ponerle un título y nombres a los ejes del gráfico. Finalmente también es posible organizar a nuestro gusto el orden de las barras, cambiando las posiciones en la lista creada.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Continentes = ['Asia', 'África', 'Europa', 'Américas', 'Oceanía']
Observaciones = [396,624,360,300,24]
ax.set_ylabel('Observaciones', color = 'red')
ax.set_xlabel('Continentes', color = 'red')
ax.set_title('Número de observaciones por continente', color = 'green')
ax.bar(Continentes,Observaciones, color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.show()
```

### Gráfico de barras horizontales
En algunos casos por restricciones de espacio o estética, entre otras razones, es ideal general el gráfico de barras acostadas o de manera horizontal. Para ello, se hace necesario realizar cambios o un código casi que totalmente diferente, tal como se ve a continuación:

```python
Continentes = ['Asia', 'África', 'Europa', 'Américas', 'Oceanía']
y_pos = np.arange(len(Continentes))
Observaciones = [396,624,360,300,24]
plt.barh(y_pos, Observaciones, align='center', color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.yticks(y_pos, Continentes)
plt.xlabel('Observaciones')
plt.title('Número de observaciones por continente')
plt.show()
```
