# Gráfico de dispersión

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calcular varianzas
varianza_gdp = round(gapminder['gdpPercap'].var())
varianza_life = round(gapminder['lifeExp'].var())

print(f"Varianza del PIB per cápita: {varianza_gdp}")
print(f"Varianza de la Expectativa de Vida: {varianza_life}")

# Crear figura con 2 subgráficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: PIB per cápita
ax1.scatter(range(len(gapminder)), gapminder['gdpPercap'], alpha=0.6, color='steelblue')
ax1.set_xlabel('Índice de Observaciones')
ax1.set_ylabel('PIB per cápita (USD)')
ax1.set_title(f'Distribución del PIB per cápita\nVarianza: {varianza_gdp}')
ax1.grid(True, alpha=0.3)

# Gráfico 2: Expectativa de Vida
ax2.scatter(range(len(gapminder)), gapminder['lifeExp'], alpha=0.6, color='coral')
ax2.set_xlabel('Índice de Observaciones')
ax2.set_ylabel('Expectativa de Vida (años)')
ax2.set_title(f'Distribución de la Expectativa de Vida\nVarianza: {varianza_life}')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
