## Personalización básica con matplotlib, según IA

```python
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder

# Configurar estilo profesional
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# Crear histogramas con personalización
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Distribución de Variables - Dataset Gapminder', 
             fontsize=16, fontweight='bold', y=0.95)

# Variables a plotear
variables = ['pop', 'lifeExp', 'gdpPercap']

for i, var in enumerate(variables):
    # Histograma superior
    axes[0, i].hist(gapminder[var], bins=30, alpha=0.7, edgecolor='black', linewidth=0.5)
    axes[0, i].set_title(f'Distribución de {var}', fontweight='bold')
    axes[0, i].set_xlabel(var)
    axes[0, i].set_ylabel('Frecuencia')
    axes[0, i].grid(True, alpha=0.3)
    
    # Boxplot inferior
    axes[1, i].boxplot(gapminder[var].dropna(), vert=False)
    axes[1, i].set_title(f'Boxplot de {var}', fontweight='bold')
    axes[1, i].set_xlabel(var)

plt.tight_layout()
plt.show()
```

## Personalización del histograma con la librería seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder

# Configuración de estilo
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# Crear subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Histogramas con KDE
sns.histplot(data=gapminder, x='lifeExp', kde=True, ax=axes[0,0], 
             color='skyblue', edgecolor='black', alpha=0.7)
axes[0,0].set_title('Expectativa de Vida', fontweight='bold')

sns.histplot(data=gapminder, x='gdpPercap', kde=True, ax=axes[0,1],
             color='lightcoral', edgecolor='black', alpha=0.7)
axes[0,1].set_title('PIB per cápita', fontweight='bold')

sns.histplot(data=gapminder, x='pop', kde=True, ax=axes[1,0],
             color='lightgreen', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Población', fontweight='bold')
axes[1,0].set_xscale('log')  # Escala logarítmica para mejor visualización

# Heatmap de correlación
corr = gapminder[['lifeExp', 'gdpPercap', 'pop']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, ax=axes[1,1])
axes[1,1].set_title('Matriz de Correlación', fontweight='bold')

plt.suptitle('Análisis Exploratorio - Dataset Gapminder', 
             fontsize=16, fontweight='bold', y=0.95)
plt.tight_layout()
plt.show()
```

## Histogramas por continente sugiere IA

```python
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder

# Configuración
sns.set_palette("Set2")
plt.figure(figsize=(15, 10))

# Histogramas separados por continente
plt.subplot(2, 2, 1)
for continent in gapminder['continent'].unique():
    data = gapminder[gapminder['continent'] == continent]['lifeExp']
    sns.histplot(data, kde=True, alpha=0.5, label=continent)
plt.title('Expectativa de Vida por Continente', fontweight='bold')
plt.legend()

plt.subplot(2, 2, 2)
for continent in gapminder['continent'].unique():
    data = gapminder[gapminder['continent'] == continent]['gdpPercap']
    sns.histplot(data, kde=True, alpha=0.5, label=continent)
plt.title('PIB per cápita por Continente', fontweight='bold')
plt.legend()
plt.xscale('log')

plt.subplot(2, 2, 3)
for continent in gapminder['continent'].unique():
    data = gapminder[gapminder['continent'] == continent]['pop']
    sns.histplot(data, kde=True, alpha=0.5, label=continent)
plt.title('Población por Continente', fontweight='bold')
plt.legend()
plt.xscale('log')

plt.subplot(2, 2, 4)
sns.boxplot(data=gapminder, x='continent', y='lifeExp')
plt.title('Boxplot - Expectativa de Vida por Continente', fontweight='bold')
plt.xticks(rotation=45)

plt.suptitle('Análisis por Continentes - Dataset Gapminder', 
             fontsize=16, fontweight='bold', y=0.95)
plt.tight_layout()
plt.show()
```

## Versión más compacta y profesional según IA

```python
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder

# Configuración profesional
sns.set_style("ticks")
plt.rcParams['figure.figsize'] = (14, 10)
plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.titleweight'] = 'bold'

# Crear figura
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

variables = ['lifeExp', 'gdpPercap', 'pop']
titles = ['Expectativa de Vida', 'PIB per cápita (log)', 'Población (log)']

for i, (var, title) in enumerate(zip(variables, titles)):
    # Histograma con KDE
    sns.histplot(data=gapminder, x=var, kde=True, ax=axes[0, i],
                 color=sns.color_palette("husl")[i], 
                 edgecolor='white', linewidth=0.5)
    axes[0, i].set_title(title)
    axes[0, i].set_xlabel('')
    
    if var in ['gdpPercap', 'pop']:
        axes[0, i].set_xscale('log')
    
    # Boxplot
    sns.boxplot(data=gapminder, y=var, ax=axes[1, i],
                color=sns.color_palette("husl")[i])
    axes[1, i].set_title(f'Boxplot - {title.split("(")[0]}')
    axes[1, i].set_xlabel('')
    
    if var in ['gdpPercap', 'pop']:
        axes[1, i].set_yscale('log')

plt.suptitle('Análisis Profesional de Distribuciones - Gapminder Dataset',
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```

## Sugerencia 1: Histograma con KDE y estadísticas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from gapminder import gapminder
from scipy import stats

# Configuración profesional
plt.style.use('default')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# Crear figura con subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Histograma con KDE en escala logarítmica
pop_log = np.log10(gapminder['pop'].dropna())
sns.histplot(pop_log, kde=True, ax=ax1, color='teal', 
             edgecolor='white', linewidth=0.5, alpha=0.8)
ax1.set_xlabel('Población (log₁₀)')
ax1.set_ylabel('Frecuencia')
ax1.set_title('Distribución de Población (Escala Logarítmica)', 
              fontweight='bold', pad=20)
ax1.grid(True, alpha=0.3)

# Añadir líneas de estadísticas descriptivas
mean_log = pop_log.mean()
median_log = pop_log.median()
ax1.axvline(mean_log, color='red', linestyle='--', linewidth=2, 
            label=f'Media: {10**mean_log:,.0f}')
ax1.axvline(median_log, color='green', linestyle='--', linewidth=2, 
            label=f'Mediana: {10**median_log:,.0f}')
ax1.legend()

# Boxplot en escala logarítmica
sns.boxplot(y=pop_log, ax=ax2, color='lightblue')
ax2.set_ylabel('Población (log₁₀)')
ax2.set_title('Boxplot de Población', fontweight='bold', pad=20)
ax2.grid(True, alpha=0.3)

# Convertir ticks a formato legible
for ax in [ax1, ax2]:
    ticks = ax.get_xticks() if ax == ax1 else ax.get_yticks()
    labels = [f'$10^{{{int(x)}}}$' for x in ticks]
    if ax == ax1:
        ax.set_xticklabels(labels)
    else:
        ax.set_yticklabels(labels)

plt.suptitle('Análisis Profesional de la Distribución de Población - Gapminder', 
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

# Estadísticas descriptivas
print("Estadísticas descriptivas de Población:")
print(f"Media: {gapminder['pop'].mean():,.0f}")
print(f"Mediana: {gapminder['pop'].median():,.0f}")
print(f"Mínimo: {gapminder['pop'].min():,.0f}")
print(f"Máximo: {gapminder['pop'].max():,.0f}")
print(f"Desviación estándar: {gapminder['pop'].std():,.0f}")
```
